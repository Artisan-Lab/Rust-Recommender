```python
# 对每页的所有问答返回超级链接以及tags
def parse_tagged_pages(html):
    question_summaries = html.find(".s-post-summary.js-post-summary")
    page_datas = []
    for q_el in question_summaries:
        question_data = {}
        #对每个questions 获取其中的hyplerlink
        hyper_link = q_el.find('div.s-post-summary--content > h3 > a', first = True)
        
        tags = q_el.find('div.s-post-summary--meta > div.s-post-summary--meta-tags', first=True)
        question_data['hyper_link'] = list(hyper_link.links)[0]
        question_data['tags'] = tags.text.replace('\n', ' ')
        page_datas.append(question_data)
    return page_datas
```

```python
def extract_data_from_url(url):
    r = requests.get(url)
    if r.status_code not in range(200, 299):
        return {}
    html_str = r.text
    html = HTML(html = html_str)
    page_datas = parse_tagged_pages(html)
    return page_datas
```
